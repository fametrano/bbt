```python
import pandas as pd
pd.set_option('display.max_rows', 100)
```

```python
def field_table(p):
    elements = range(p)
    
    opposites = []
    for element in elements:
        opposites.append((p-element)%p)

    inverses = ['nan'] + [i**(p-2) % p for i in elements[1:]]
    
    odd_sqrts = [0]
    even_sqrts = [0]
    for element in range(1, p): # doppio loop, si puo` ottimizzare
        skip=False
        for i in range(1, p, 2): # numeri dispari
            if i**2 % p == element:
                odd_sqrts.append(i)
                even_sqrts.append(p-i)
                skip=True
                break
        if not skip:
            odd_sqrts.append('nan')
            even_sqrts.append('nan')
            
    table = pd.DataFrame()
    table['opposite'] = opposites
    table['inverse'] = inverses
    table['odd sqrt'] = odd_sqrts
    table['even sqrt'] = even_sqrts
    return table
```
